```python
import numpy as np
from scipy.stats import norm
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
def true_sigma(th):
    return 1+th**2

def coeffs(th):
    return np.asarray([1,th,th**2])

def coeff_inverse(basis):
    m = []
    for elem in basis:
        m.append(coeffs(elem))
    return np.linalg.inv(m)


```

```python
for j in range(3):

    thetas = np.linspace(-30,30,100)
    true = true_sigma(thetas)
    shifted_true = true*np.random.normal(1,0.001,100)
    plt.plot(thetas,(shifted_true-true)/true,label="Point shift")
    basis = np.random.uniform(-30,30,3)
    if j == 0:
        basis = np.asarray([-2,0,2])
    for h in range(1000):
        res = np.random.normal(1,0.001,3)
        benches = true_sigma(basis)*res
        predict = []
        for elem in thetas:
            predict.append(coeffs(elem)@coeff_inverse(basis)@benches)
        plt.plot(thetas,(predict-true)/true,alpha=0.02)
    print(basis)
    plt.legend()
    plt.show()
    plt.

    plt.clf()
```
